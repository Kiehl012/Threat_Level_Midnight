```python
# Dependencies
import pandas as pd
import numpy as np
 
```

```python
# Open cleaned script
file_path ='Data/cleaned_scripts.csv'
df = pd.read_csv(file_path,index_col=0)
df.head(10)
```

```python
bloopers_df = df[df['episode'] ==99]
bloopers_df.head()
```

```python
# Remove rows for episodes 99
df.drop(df[df['episode'] == 99].index, inplace = True)
df.head(10)
```

```python
# Merged Cleaned Data with Episode Details csv
file_path ='Data/Episode_Detail.csv'
df_ed = pd.read_csv(file_path,index_col=False)
merged_df = df.merge(df_ed, left_on='line_ID', right_on='Line_ID')
merged_df = merged_df.drop(columns=['season', 'episode', 'line_ID'])
merged_df.tail(50)
```

```python
# line count
lines_df = merged_df.groupby(['Line_ID','Original air date','character'])['line'].count().reset_index(name='line_count').sort_values(['character','Line_ID'])
lines_df
```

```python
# Get list of all characters
character_list = lines_df['character'].unique().tolist()
character_list.sort()
print(character_list)
```

```python
character_df = pd.DataFrame()
character_df['character']=character_list
character_df
```

```python
character_x_df = pd.concat([character_df]*186, ignore_index=True)
character_x_df = character_x_df.sort_values('character')
character_x_df
```

```python
# Export to csv (is in order andy, andy andy...)
character_x_df.to_csv('./Data/character.csv', index=False)
```

```python
# Get list of all line_id
line_ID_list = lines_df['Line_ID'].unique().tolist()
line_ID_list.sort()
print(line_ID_list)
```

```python
line_ID_df = pd.DataFrame()
line_ID_df['Line_ID']=line_ID_list
line_ID_df
```

```python
line_ID_x_df = pd.concat([line_ID_df]*32, ignore_index=False)
#line_ID_x_df = line_ID_x_df.sort_values('Line_ID')
line_ID_x_df
```

```python
line_ID_x_df.to_csv('./Data/line_ID.csv', index=False)
```

```python
import pandas as pd
```

```python
test_df = pd.read_csv("cleaned_scripts.csv", index_col=0)
test_df.head(20)
```

```python
test_df.loc[(test_df['character']=='Dwight') & (test_df['season']==1), 'character'].count()
```

```python
# create new Dataframe for total lines per character per season
total_lines_df = test_df.groupby(['season', 'character'])['character'].count().reset_index(name="lines_per_season")
total_lines_df
```

```python
import datetime
# adding arbitrary year to each season
first_air_date = datetime.date(2003, 3, 24)
total_lines_df['arb_date'] = first_air_date
total_lines_df.head()
```

```python
total_lines_df['date'] = total_lines_df[['arb_date', 'season']].apply(lambda x: x['arb_date'] + 
                                                                  pd.DateOffset(years=x['season']), axis=1)
total_lines_df.head()
```

```python
total_lines_df.tail()
```

```python
# create new DataFrame for total lines per character per season
total_lines2_df = pd.pivot_table(test_df, 
               index=['season', 'character'],
               aggfunc={'character': len}).rename(columns={'character': 'total_lines'})
total_lines2_df
```

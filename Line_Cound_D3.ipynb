```python
import pandas as pd
import numpy as np
 
```

```python
file_path ='Data/cleaned_scripts.csv'
df = pd.read_csv(file_path,index_col=0)
df.head(40)
```

```python
df = df[['character', 'line','line_ID']]
df.head(40)
```

```python
line_count = df.groupby([ 'line_ID', 'character']).count().sort_values(['line_ID'])
line_count.head(40)
```

```python
line_count.to_csv('./Data/line_count.csv')
```

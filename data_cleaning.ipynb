```python
# Find Characters with >100 lines and create new df that includes only those characters
character_list = lines.query('line>100')['character'].to_list()
new_df = df[df['character'].isin(character_list)]
new_df
```

```python
# Check df for stage directions and other non-script text as defined by text inside of square brackets '[]'
len(new_df[new_df['line'].str.contains('\]|\[')])
```

```python
# Removed stage directions and other non-script text as defined by text instead of square brackets '[]'
new_df['line'] = new_df['line'].str.replace(r"\[[^()]*\]", "", regex=True)
len(new_df[new_df['line'].str.contains('\]|\[')])
```

```python
# Convert blank lines to NaN and remove rows with NaN values.
new_df['line'] = new_df['line'].str.strip()
new_df = new_df.replace('', np.NaN, regex=False)
new_df = new_df.dropna()
```

```python
line = new_df['line'].tolist()
```

```python
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyser = SentimentIntensityAnalyzer()
```

```python
line_scores = []
for sentence in line:
    score = analyser.polarity_scores(sentence)
    line_scores.append(score)
#Converting List of Dictionaries into Dataframe
lineScore_df= pd.DataFrame(line_scores)
lineScore_df.head(20)
```

```python
sentiment = lineScore_df['compound'].tolist()
```

```python
new_df['sentiment_score'] = sentiment
new_df.head(20)
```

```python
# export new_df to csv file
new_df.to_csv('Data/cleaned_scripts.csv')
```

```python
cleanline = new_df['line'].tolist()
```

```python
from spellchecker import SpellChecker

spell = SpellChecker() #loads default word frequency list
```

```python
# find those words that may be misspelled
misspelled = spell.unknown(cleanline)

for word in misspelled:
    # Get the one `most likely` answer
    print(spell.correction(word))

    # Get a list of `likely` options
    print(spell.candidates(word))
```
